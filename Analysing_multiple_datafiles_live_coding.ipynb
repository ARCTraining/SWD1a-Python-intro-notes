```python
# let's build a plot for inflam-01

filename = filenames_subset[0]
print("Filename: "+ filename)

dataset_name = filename[-19:-4]
print("Dataset: " + dataset_name)

# load in the data
data = np.loadtxt(fname=filename, delimiter=",")

# set up the figure

fig = plt.figure(figsize=(5.0, 7.0), layout="compressed")

# set up individual axes
axes1 = fig.add_subplot(2, 1, 1)
axes2 = axes1.twinx()
axes3 = fig.add_subplot(2, 1, 2)

# plot some data
# top panel
axes1.plot(np.mean(data, axis=0), c="black")
axes2.plot(np.amin(data, axis=0), c="grey", linestyle="--")
axes1.set_xlabel("Days")
axes1.set_ylabel("Mean inflammation")
axes2.set_ylabel("Min. inflammation", c="grey")

#lower panel
heatmap = axes3.imshow(data, aspect="auto")
cbar = fig.colorbar(heatmap, location="bottom")
cbar.set_label("Inflammation")
axes3.set_xlabel("Days")
axes3.set_ylabel("Patient ID #")

axes1.set_xlim(-0.5, 39.5)
axes3.set_xlim(-0.5, 39.5)

fig.suptitle("Results of dataset: "+ dataset_name)

save_path = "/content/swc-python/plots/"

plt.savefig(save_path + dataset_name + ".png")
```

```python
#  lets turn this into a loop

for filename in filenames_subset:
    print("Filename: "+ filename)

    dataset_name = filename[-19:-4]
    print("Dataset: " + dataset_name)

    # load in the data
    data = np.loadtxt(fname=filename, delimiter=",")

    # set up the figure

    fig = plt.figure(figsize=(5.0, 7.0), layout="compressed")

    # set up individual axes
    axes1 = fig.add_subplot(2, 1, 1)
    axes2 = axes1.twinx()
    axes3 = fig.add_subplot(2, 1, 2)

    # plot some data
    # top panel
    axes1.plot(np.mean(data, axis=0), c="black")
    axes2.plot(np.amin(data, axis=0), c="grey", linestyle="--")
    axes1.set_xlabel("Days")
    axes1.set_ylabel("Mean inflammation")
    axes2.set_ylabel("Min. inflammation", c="grey")

    #lower panel
    heatmap = axes3.imshow(data, aspect="auto")
    cbar = fig.colorbar(heatmap, location="bottom")
    cbar.set_label("Inflammation")
    axes3.set_xlabel("Days")
    axes3.set_ylabel("Patient ID #")

    axes1.set_xlim(-0.5, 39.5)
    axes3.set_xlim(-0.5, 39.5)

    fig.suptitle("Results of dataset: "+ dataset_name)

    save_path = "/content/swc-python/plots/"

    plt.savefig(save_path + dataset_name + ".png")
```

```python
# to compare datasets

data0 = np.loadtxt(sorted_filenames[0], delimiter=",")
data1 = np.loadtxt(sorted_filenames[1], delimiter=",")

fig = plt.figure(figsize=(10.0, 3.0))

plt.plot(np.mean(data0, axis=0) - np.mean(data1, axis=0))
plt.ylabel("Difference in average inflammation")
plt.xlabel("Days")
plt.tight_layout()
```

```python
# create an array filled with zeroes
composite_data = np.zeros((60, 40))

# now we want to get the average of all files
# lets do this by adding the arrays together (elementwise)
# then dive the end result by the number of frames

for filename in sorted_filenames:
    data = np.loadtxt(fname = filename, delimiter=",")
    composite_data = composite_data + data
```

```python
print(composite_data)
```

```python
composite_data = composite_data/len(sorted_filenames)
```

```python

dataset_name = "Composite"
print("Dataset: " + dataset_name)

# load in the data

data = composite_data
# set up the figure

fig = plt.figure(figsize=(5.0, 7.0), layout="compressed")

# set up individual axes
axes1 = fig.add_subplot(2, 1, 1)
axes2 = axes1.twinx()
axes3 = fig.add_subplot(2, 1, 2)

# plot some data
# top panel
axes1.plot(np.mean(data, axis=0), c="black")
axes2.plot(np.amin(data, axis=0), c="grey", linestyle="--")
axes1.set_xlabel("Days")
axes1.set_ylabel("Mean inflammation")
axes2.set_ylabel("Min. inflammation", c="grey")

#lower panel
heatmap = axes3.imshow(data, aspect="auto")
cbar = fig.colorbar(heatmap, location="bottom")
cbar.set_label("Inflammation")
axes3.set_xlabel("Days")
axes3.set_ylabel("Patient ID #")

axes1.set_xlim(-0.5, 39.5)
axes3.set_xlim(-0.5, 39.5)

fig.suptitle("Results of dataset: "+ dataset_name)

# save_path = "/content/swc-python/plots/"

# plt.savefig(save_path + dataset_name + ".png")
```
